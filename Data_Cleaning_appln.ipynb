```python
# Data Cleaning Master - Python Application

# This is a data cleaning application 

"""
Please create a python application that can take datasets and clean the data
- It should ask for datasets path and name
- it should check number of duplicats and remove all the duplicates 
- it should keep a copy of all the duplicates
- it should check for missing values 
- if any column that is numeric it should replace nulls with mean else it should drop that rows
- at end it should save the data as clean data and also return 
- duplicates records, clean_data 
"""

# to check the virtual environment
!pip list

#install & import dependencies

!pip install openpyxl

!pip install xlrd

import pandas as pd
import numpy as np
import time
import openpyxl
import xlrd
import os
import random

#df = pd.read_excel("C:\\Shanthi\\Data_Analyst\\Python\\python-Data_Cleaning_Application\\Datasets\\walmart\\walmart.xlsx")

def data_cleaningApp(data_path,data_name):
    
    print("Thank you for uploading dataset...checking the datafile")

    #Generating random numbers
    sec = random.randint(1,4)

    #printing delay messages...
    print(f" \n Please wait for {sec} sec... data path is getting validated...")
    time.sleep(sec)
    
    #df = pd.read_excel("C:\\Shanthi\\Data_Analyst\\Python\\python-Data_Cleaning_Application\\Datasets\\walmart\\walmart.xlsx")
    
# task1 - - It should ask for datasets path and name
    
    # get data path and dataset name
        #data_path = "C:\\Shanthi\\Data_Analyst\\Python\\python-Data_Cleaning_Application\\Datasets\\walmart\\walmart.xlsx"
        #data_name = "walmart"

    #check if the given path is valid or not and check the data_type of given file
    
    if not os.path.exists(data_path):
        print("Incorrect path...Try with the correct path")
        return
    else:
        if data_path.endswith("csv"):
            print("Its a CSV file..")
            data = pd.read_csv(data_path)
        elif data_path.endswith("xlsx"):
            print("Its a excel file..")
            data = pd.read_excel(data_path)
        else:
            print("Invalid datatype..")
            return

    # showing number of records 

    #Generating random numbers
    sec = random.randint(1,4)
    #printing delay messages...
    print(f"Please wait for {sec} sec... Fetching count of data rows and columns...")
    time.sleep(sec)
    
    #method1
    #num_rows = data.shape
    #print("Number of columns and rows in the given dataset :",num_rows)
    
    #method2
    print(f'Number of columns and rows in the given dataset... \n Rows: {data.shape[0]} \n Columns: {data.shape[1]}')
    
# Task2 - - it should check number of duplicats and remove all the duplicates 
    
    # Start Cleaning

    #Generating random numbers
    sec = random.randint(1,4)
    #printing delay messages...
    print(f"Please wait for {sec} sec... Checking for duplicates...")
    time.sleep(sec)
    
    # find duplicates
    
    duplicates = data.duplicated()
    total_duplicates = data.duplicated().sum()
    print(f"Total count of Duplicated data in the given file : {total_duplicates}")


    #print(duplicate_record)
    
    # saving duplicate record in the file directory
    
    if not total_duplicates==0:
        duplicate_record = data[duplicates]
        duplicate_record.to_csv(f'{data_name}_duplicates.csv', index = None)
        print("Duplicates found..")
    else :
        print("No duplicates found...")
    
    
    # deleting duplicates and saving unique records in the file directory
    
    df = data.drop_duplicates()
    df.to_csv(f'{data_name}_unique_records.csv', index=None)
    
# Task3 --it should check for missing values 

    #Generating random numbers
    sec = random.randint(1,4)
    print(f"Please for {sec} sec ...Checking for missing values...")
    time.sleep(sec)
    
    # it will return the total count of missing values
    total_missing_values = df.isnull().sum().sum()
    print("Total number of missing values : ",total_missing_values)

    # it will return the total count of missing values by each columns
    missing_values_bycolumn = df.isnull().sum()
    print("Total number of missing values by each columns: ",missing_values_bycolumn)
    
#Task4 --if any column that is numeric it should replace nulls with mean else it should drop that rows
    
    # Check datatypes
    column = df.columns
    print("Contain columns are: ",column)
    print("Given columns datatypes are :",df.dtypes)
    
    #looping through a columns to check data types of each
    # replace to mean if the column datatype is int and float
    # Notes - here using 
           # fillna - if columns has int and float
            #dropna - for rest of all data_types (other than int,float)
    
    for col in column:
        if df[col].dtypes in (int, float):
            print("Its int,float")
            df[col] = df[col].fillna(df[col].mean())
        else:
            df.dropna(subset=col, inplace=True)
    
    print("Data is cleaned now...")



    # Getting shape of file after handling duplicates
    
    print(f'Number of dataset shape after cleaned... rows : {df.shape[0]} and columns : {df.shape[1]}')
    
#Task5 --at end it should save the data as clean data and also return
    
    #Storing the cleaned data in file directory
    
    df.to_csv(f'{data_name}_cleaned_data.csv', index=None)
    print("Dataset is saved..")


if__name__ == "__Data_Cleaning_appln__":

print("Welcome to the Data cleaning Master")
data_path = input("Please give datapath : ")
data_name = input("Please give dataset name : ")
#data_cleaningApp("C:\\Shanthi\\Data_Analyst\Python\\python-Data_Cleaning_Application\\Datasets\\walmart\\walmart.xlsx","walmart")
data_cleaningApp(data_path,data_name)
```
